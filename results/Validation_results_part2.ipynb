Refit top models on TRAIN+VAL, evaluate on Test (with clone)

```python
from sklearn.base import clone

# Choose ranking metric: macro F1 (per professor suggestion)
results_val_sorted = results_val.sort_values("Val_F1_macro", ascending=False)

print("\nTop models by validation Macro F1:")
print(results_val_sorted.head(2))

top_models = results_val_sorted["Model"].head(2).tolist()

def get_model_and_data(name):
    if name == "DT + RUS":
        model = best_rus_dt
        X_train_full = pd.concat([X_train, X_val], axis=0)
        y_train_full = pd.concat([y_train, y_val], axis=0)
        X_test_final = X_test
        y_test_final = y_test
        is_xgb = False

    elif name == "RF + RUS":
        model = best_rus_rf
        X_train_full = pd.concat([X_train, X_val], axis=0)
        y_train_full = pd.concat([y_train, y_val], axis=0)
        X_test_final = X_test
        y_test_final = y_test
        is_xgb = False

    elif name == "XGB + RUS":
        model = best_rus_xgb
        X_train_full = pd.concat([X_train_xgb, X_val_xgb], axis=0)
        y_train_full = np.concatenate([y_train_xgb, y_val_xgb])
        X_test_final = X_test_xgb
        y_test_final = y_test   # original labels for reporting
        is_xgb = True

    elif name == "DT + BalancedBagging":
        model = best_bb_dt
        X_train_full = pd.concat([X_train, X_val], axis=0)
        y_train_full = pd.concat([y_train, y_val], axis=0)
        X_test_final = X_test
        y_test_final = y_test
        is_xgb = False

    elif name == "RF + BalancedBagging":
        model = best_bb_rf
        X_train_full = pd.concat([X_train, X_val], axis=0)
        y_train_full = pd.concat([y_train, y_val], axis=0)
        X_test_final = X_test
        y_test_final = y_test
        is_xgb = False

    elif name == "XGB + BalancedBagging":
        model = best_bb_xgb
        X_train_full = pd.concat([X_train_xgb, X_val_xgb], axis=0)
        y_train_full = np.concatenate([y_train_xgb, y_val_xgb])
        X_test_final = X_test_xgb
        y_test_final = y_test   # original labels for reporting
        is_xgb = True

    else:
        raise ValueError(f"Unknown model name: {name}")

    return model, (X_train_full, y_train_full, X_test_final, y_test_final), name, is_xgb

for name in top_models:
    model, (X_train_full, y_train_full, X_test_final, y_test_final), label, is_xgb = get_model_and_data(name)

    print(f"\n\n===== Re-fitting {label} on TRAIN+VAL and evaluating on TEST =====")
    fresh_model = clone(model)
    fresh_model.fit(X_train_full, y_train_full)

    _ = evaluate_classifier(
        fresh_model,
        X_test_final,
        y_test_final,
        label=f"{label} (TEST)",
        average="weighted",
        is_xgb=is_xgb
    )
```

# Class weight variants

```python
# ============================================================
# 9. Class-weighted models (no resampling)
# ============================================================

# ---- Decision Tree with class_weight="balanced" ----
dt_weighted = DecisionTreeClassifier(
    random_state=42,
    class_weight="balanced"
)

param_dist_dt_w = {
    "max_depth": [None, 5, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

best_dt_weighted = randomized_search(
    dt_weighted,
    param_dist_dt_w,
    X_train,
    y_train,
    n_iter=10
)

dt_weighted_metrics = evaluate_classifier(
    best_dt_weighted,
    X_val,
    y_val,
    label="DT + class_weight=balanced"
)

# ---- Random Forest with class_weight="balanced" ----
rf_weighted = RandomForestClassifier(
    random_state=42,
    n_jobs=-1,
    class_weight="balanced"
)

param_dist_rf_w = {
    "n_estimators": [100, 200, 300],
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2]
}

best_rf_weighted = randomized_search(
    rf_weighted,
    param_dist_rf_w,
    X_train,
    y_train,
    n_iter=10
)

rf_weighted_metrics = evaluate_classifier(
    best_rf_weighted,
    X_val,
    y_val,
    label="RF + class_weight=balanced"
)
```

```python
# ============================================================
# 10. XGBoost with class-based sample weights
# ============================================================

# Compute class weights like sklearn's "balanced"
class_counts = y_train.value_counts().sort_index()  # labels 1..7
total_samples = len(y_train)
n_classes = class_counts.shape[0]
class_weights = {
    cls: total_samples / (n_classes * count)
    for cls, count in class_counts.items()
}

# Map each training sample to its class weight
sample_weights_train = y_train.map(class_weights).values

xgb_weighted = XGBClassifier(
    objective="multi:softprob",
    eval_metric="mlogloss",
    random_state=42,
    n_jobs=-1
)

param_dist_xgb_w = {
    "n_estimators": [200, 300, 400],
    "learning_rate": [0.05, 0.1, 0.2],
    "max_depth": [4, 6, 8],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0]
}

# We can't pass sample_weight directly into RandomizedSearchCV params,
# so we do manual tuning or a simpler search. To keep it consistent
# with your other code, we can pick a reasonable fixed configuration
# or do a very small manual loop. For simplicity, here we pick one
# configuration and focus on the weighting:

best_xgb_weighted = XGBClassifier(
    objective="multi:softprob",
    eval_metric="mlogloss",
    random_state=42,
    n_jobs=-1,
    n_estimators=300,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8
)

# Fit on label-encoded y (0..6) but with sample weights based on original y_train
best_xgb_weighted.fit(
    X_train_xgb,
    y_train_xgb,
    sample_weight=sample_weights_train
)

xgb_weighted_metrics = evaluate_classifier(
    best_xgb_weighted,
    X_val_xgb,
    y_val,                 # original labels for reporting
    label="XGB + sample weights",
    is_xgb=True
)
```
