# Class-imbalance Techniques

## Random Undersampling

```python
rus = RandomUnderSampler(random_state=42)

# ---- Decision Tree + RUS ----
dt_base = DecisionTreeClassifier(random_state=42)
rus_dt_pipe = ImbPipeline([
    ("rus", rus),
    ("clf", dt_base)
])

param_dist_dt = {
    "clf__max_depth": [None, 5, 10, 20, 30],
    "clf__min_samples_split": [2, 5, 10],
    "clf__min_samples_leaf": [1, 2, 4]
}

best_rus_dt = randomized_search(
    rus_dt_pipe,
    param_dist_dt,
    X_train,
    y_train,
    n_iter=10
)

rus_dt_metrics = evaluate_classifier(
    best_rus_dt,
    X_val,
    y_val,
    label="RUS + Decision Tree"
)

# ---- Random Forest + RUS ----
rf_base = RandomForestClassifier(random_state=42, n_jobs=-1)
rus_rf_pipe = ImbPipeline([
    ("rus", rus),
    ("clf", rf_base)
])

param_dist_rf = {
    "clf__n_estimators": [100, 200, 300],
    "clf__max_depth": [None, 10, 20],
    "clf__min_samples_split": [2, 5],
    "clf__min_samples_leaf": [1, 2]
}

best_rus_rf = randomized_search(
    rus_rf_pipe,
    param_dist_rf,
    X_train,
    y_train,
    n_iter=10
)

rus_rf_metrics = evaluate_classifier(
    best_rus_rf,
    X_val,
    y_val,
    label="RUS + Random Forest"
)

# ---- XGBoost + RUS ----
xgb_base = XGBClassifier(
    objective="multi:softprob",
    eval_metric="mlogloss",
    random_state=42,
    n_jobs=-1
)

rus_xgb_pipe = ImbPipeline([
    ("rus", rus),
    ("clf", xgb_base)
])

param_dist_xgb = {
    "clf__n_estimators": [200, 300, 400],
    "clf__learning_rate": [0.05, 0.1, 0.2],
    "clf__max_depth": [4, 6, 8],
    "clf__subsample": [0.8, 1.0],
    "clf__colsample_bytree": [0.8, 1.0]
}

best_rus_xgb = randomized_search(
    rus_xgb_pipe,
    param_dist_xgb,
    X_train_xgb,
    y_train_xgb,
    n_iter=10
)

rus_xgb_metrics = evaluate_classifier(
    best_rus_xgb,
    X_val_xgb,
    y_val,              # original labels for reporting
    label="RUS + XGBoost",
    is_xgb=True
)
```
